```python
import numpy as np
import matplotlib.pyplot as plt
```

This program solves for the Mandelbrot Set. The equation is relies on is

$$z' = z^{2} + c$$

For $|z| < 2$

```python
#Initial Parameters
# z is the initial complex equation
# N_iter is the number of times the equation z = z^2 + c is to be solved as a maximum
# N is the number of values on each axis
# min_val, max_val are the range

z = 0 + 0j
N_iter = 100
N = 1000
min_val = -2
max_val = 2

# Calculate Steps
h = (max_val - min_val)/N

# Create empty array
mandel = np.empty([N, N], float)

# Solve for z, c, using the min_vals and max_vals
for i in range(len(mandel[0])):
    for j in range(len(mandel)):
        z = complex(0,0)
        x = min_val + (h*i)
        y = min_val + (h*j)
        c = complex(x, y)
        for k in range(N_iter):
            if abs(z) > 2:
                break
            else:
                z = (z**2) + c
        mandel[i][j] = abs(z)

# Plotting
plt.imshow(mandel)
plt.title("Mandelbrot Set")
plt.xlabel("Re(c)")
plt.ylabel("Im(c)")
plt.gray()
cb = plt.colorbar()
cb.ax.set_ylabel('|z|')
plt.show()
```
